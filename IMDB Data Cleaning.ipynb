```
movies_dict = {}

for value in imdb_data_movies['imdb_id']:
  try:
    movie_id = value
    movies_dict[movie_id] = {}
  except:
    pass
```

```
len(movies_dict)
```

# genres

```
imdb_data_movies['genres'].unique()
```

```
genre_ids = {}
```

```
import ast

for m, b in zip(movies_dict, imdb_data_movies['genres']):
  z=ast.literal_eval(b)

  #loop over every genre in list of genres within one movie
  for index, i in enumerate(z):
    id = i['id']
    name = i['name']
    genre_ids[id] = name
```

```
genre_ids
```

```
ids_of_wrong_genres = [7761,2883,7759,7760,23822, 7761,11176,11602,11602,17161,18012,18013,18013,29812,33751]
```

```
for id in ids_of_wrong_genres:
  genre_ids.pop(id, None)
```

```
genre_ids
```

```
len(movies_dict)
```

```
import ast
to_delete = []
for m, b in zip(movies_dict, imdb_data_movies['genres']):
  z=ast.literal_eval(b)

  #loop over every genre in list of genres within one movie
  for index, i in enumerate(z):
    id = i['id']
    name = i['name']
    genre_ids['name'] = name
    if id in genre_ids and index < 3:
      movies_dict[m][f'genre_{index}'] = name
    else:
      print(f'Error with: {m}')
      to_delete.append(m)
```

```
for id in to_delete:
  movies_dict.pop(id, None)
```

# production_countries

```
imdb_data_movies['production_countries'].unique()
```

```
country_ids = {}

for m, b in zip(movies_dict, imdb_data_movies['production_countries']):
  try:
    z=ast.literal_eval(b)

    #loop over every genre in list of genres within one movie
    for index, i in enumerate(z):
      id = i['iso_3166_1']
      name = i['name']
      country_ids[id] = name
  except:
    pass
```

```
to_delete = []

for m, b in zip(movies_dict, imdb_data_movies['production_countries']):
  try:
    z=ast.literal_eval(b)

    #loop over every genre in list of genres within one movie
    for index, i in enumerate(z):
      id = i['iso_3166_1']
      
      if id in country_ids and index < 3:
        movies_dict[m][f'prod_country_{index}'] = id
      else:
        print(f'Error with: {m}')
        to_delete.append(m)
  except:
    print(f'Error with: {m}')
    to_delete.append(m)
```

```
for id in to_delete:
  movies_dict.pop(id, None)
```
